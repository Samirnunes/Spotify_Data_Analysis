```python
def find_feat(s):
    if '(feat.' in s:
        feat = s.split('(feat.')
        return feat[1].strip(')')
    return 'NaN'
df['feat'] = df['title'].apply(find_feat)
```

```python
df[df['feat'] != 'NaN']
```

```python
def count_feats(s):
    if s == 'NaN':
        return 0
    elif ' & ' in s:
        return 2
    else:
        return len(s.split(','))
df['number of feats'] = df['feat'].apply(count_feats)
```

```python
df['added'] = pd.to_datetime(df['added'], format = '%Y‑%m‑%d')
```

```python
df['added']
```

```python
df['added year'] = [date.year for date in df['added']]
```

```python
df['added month'] = [date.month for date in df['added']]
```

```python
df['added day'] = [date.day for date in df['added']]
```

```python
sns.countplot(data = df, x = df['artist type'])
```

```python
sns.jointplot(x = df['year released'] , y = df['dur'])
```

```python
# Há dois outliers muito distoantes: um de duração e um de ano. Vamos remover os dois para melhorar a visualização
df = df[(df['year released'] > 2000) & (df['dur'] < 600)]
```

```python
sns.jointplot(x = df['year released'] , y = df['dur'])
```

```python
#Limitei o número de plots para o jupyter conseguir processar o pairplot. Considerei, para isso, apenas as correlações mais lógicas.
sns.pairplot(df[['pop','live','spch', 'acous', 'dnce' ,'nrgy', 'val','artist type']], hue = 'artist type')
```

```python
mean_dur = df[df['artist type'] == 'Solo'].groupby('top year')['dur'].mean()
```

```python
mean_dur
```

```python
plt.plot(df['top year'].unique(), mean_dur)
plt.xlabel('top year')
plt.ylabel('mean_dur')
plt.title('Evolução do valor médio da duração das músicas solo')
plt.show()
```

```python
mean_acous = df[df['artist type'] == 'Solo'].groupby('top year')['acous'].mean()
```

```python
mean_acous
```

```python
plt.plot(df['top year'].unique(), mean_acous)
plt.xlabel('top year')
plt.ylabel('mean_acous')
plt.title('Evolução do valor médio da acústica das músicas solo')
plt.show()
```
